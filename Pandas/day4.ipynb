```python
def convertGender(cell):
    if cell=="Female":
        return 'Women'
    return cell

def convertCountry(cell):
    if cell=="United States":
        return "US"
    return cell

df = pd.read_excel("file_example.xls", "Sheet2", converters = {
'Gender': convertGender,
'Country': convertCountry})
df
```

```python
df.to_excel("newExcel.xlsx")
```

```python
df.to_excel("newExcel.xlsx", index=False)
```

```python
df.to_excel("newExcel.xlsx", sheet_name="Data")
```

```python
df.to_excel("newExcelFile.xlsx", sheet_name="Data1", index=False)
```

```python
df.to_excel("newExcelFile.xlsx", sheet_name="Data1", index=False, startrow=2, startcol=3)
```
