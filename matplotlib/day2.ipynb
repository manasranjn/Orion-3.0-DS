```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [5,10,20,15,30,35]

plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Simple Line Plot")
```

```python
plt.plot(x, y, color='g', linestyle='-.', marker='o' , linewidth=2)
plt.title("Simple Line Plot")
plt.xlabel("Temperature")
plt.ylabel("Wind Speed")
```

```python
plt.plot( [10, 20, 25, 30], label='Sales 2024', color='b')
plt.plot( [15, 18, 22, 28], label='Sales 2025', color='g')
plt.plot( [12, 10, 18, 25], label='Sales 2023', color='purple')
plt.legend()
```

```python
plt.plot( [10, 20, 25, 30], label='Sales 2024')
plt.plot( [15, 18, 22, 28], label='Sales 2025')
plt.legend(loc='upper right', shadow=True)
```

```python
plt.plot([1,2,3],[3,4,5], label='Line A')
plt.plot([1,2,3],[5,3,1], label='Line B')

plt.legend(loc='best', title='Legend Title', fontsize=10, frameon=True)
```

```python
plt.plot([1,2,3],[3,4,5], label='Line A')
plt.plot([1,2,3],[5,3,1], label='Line B')

plt.legend(loc='upper left', title='Legend Title', fontsize=10, frameon=True)
plt.grid()
```
