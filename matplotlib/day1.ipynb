```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
```

```python
plt.plot(x, y, color='g')
```

```python
plt.plot(x, y, color='crimson')
```

```python
plt.plot(x, y, color='r', linestyle='--')
```

```python
plt.plot(x, y, color='r', linestyle='-.')
```

```python
plt.plot(x, y, color='r', linestyle='dotted')
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='X')
```

```python
plt.plot(x,y, ':r+')
```

```python
plt.plot(x,y, 'bx--')
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='x', markersize=10)
```

```python
plt.plot(x, y, color='b', linestyle='-.', marker='o', markersize=10 , linewidth=5)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=1)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4, alpha=0.1)
```

```python
plt.plot(x, y, color='b', linestyle='--', marker='x' , linewidth=4)
plt.title("Lilan")
```
